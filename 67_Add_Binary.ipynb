# 67. Add Binary

Given two binary strings a and b, return their sum as a binary string.

## Example 1:

```
Input: a = "11", b = "1"
Output: "100"
```

## Example 2:

```
Input: a = "1010", b = "1011"
Output: "10101"
```

## Constraints:

```
1 <= a.length, b.length <= 104
a and b consist only of '0' or '1' characters.
Each string does not contain leading zeros except for the zero itself.
```

# Solution
We are going to solve this by bit computation. I will explain the algorithm: First we are going to check which string is longer. Depening on which string is shorter, I add `'0'`'s infront of the string which is shorter. I add the amount which will make the 2 strings equal in length. Why do this? Because I'll be looping through these string. I want to loop over both strings and the loop will go from `max(len(a),len(b))-1` to `0`. In every iteration I'll be summing up 2 numbers. Each number is an element of either `a` or `b`. Now, since binary addition is not the same as normal addition, I create a function which will help me  with this. Let me first explain how binary addition works: If we have `0` and `1`, then the sum is `1`. If we have `1+1` then the sum is `10`. In order to encorporate this remainder, I need this function called `binary`. This function will take in as arguments `n1`, `n2`, `n3`. `n1` and `n2` are the numbers of the strings. `n3` is the remainder from the last addition. So in each loop I calculate 2 things - the sum of the current 2 numbers and the remainder. Note that if we are adding `1+1+n3` where `n3=1`, then the sum is `1` and the remainder is `1`. At each iteration I concatenate the sum to `sum`. At the end, I check if I have a remainder and add `1` to `sum` if needed. After that I return the reverse of `sum`.

# Complexity
Time complexity - $O(max(n,m))$, where $n$ and $m$ are the lengths of `a` and `b` respectively. Explanation: We traverse through the longer string 1 time.

Space complexity - $O(max(n,m))$ - we create a string for the output.

````python
def addBinary(a, b):
    """
    :type a: str
    :type b: str
    :rtype: str
    """
    def binary(n1, n2, n3):
      if n1 + n2 == 0:
        return str(n3), 0
      elif n1 + n2 + n3 == 1:
        return '1', 0 
      elif n1 + n2 + n3 == 2:
        return '0', 1
      elif n1 + n2 + n3 == 3:
        return '1', 1

    if len(a) > len(b):
      c = '0'*(len(a) - len(b))
      b = c + b 
    if len(b) > len(a):
      c = '0'*(len(b) - len(a))
      a = c + a 

    sum = ''
    n = 0
    for i in range(max(len(a),len(b)) -1, -1, -1):
      l, n = binary(int(a[i]), int(b[i]), n)
      sum += l
    if n == 1:
      sum += '1' 
    return sum[::-1]
    
````
