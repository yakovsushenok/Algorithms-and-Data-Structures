# 27. Remove Element

Given an integer array `nums` and an integer `val`, remove all occurrences of `val` in `nums` **in-place**. The relative order of the elements may be changed.

Since it is impossible to change the length of the array in some languages, you must instead have the result be placed in the first part of the array `nums`. More formally, if there are `k` elements after removing the duplicates, then the first `k` elements of `nums` should hold the final result. It does not matter what you leave beyond the first `k` elements.

Return `k` after placing the final result in the first `k` slots of nums.

Do not allocate extra space for another array. You must do this by modifying the input array in-place with $O(1)$ extra memory.

# Solution

Since we want to solve this problem without creating a new arrary (i.e. constant space complexity) we will use the two-pointer technique. The first pointer `i` will be the pointer which is going to go only the number of times equal to the number of numbers which are not equal to `val` i.e., if we have 5 non-`val` and 3 `val`, then `i` is going to go only until the 4-th index. The second pointer is going to go through the whole array and it's going to check if the values are equal to `val`. For each non-`val`, `j` is going to make the value at `i`-th index change it's value to `num[j]`. Example:

```
nums = [0,1,2,2,3,0,4,2]
val = 2
i = 0
for j in [0, len(nums)]:

i = 0, j = 0, nums[i] = 0, nums[j] = 0 => nums[j]!=2 => set nums[i] = 0 and i = i+1

i = 1, j=1, nums[i] = 1, nums[j] = 1 => nums[j]!=2 => set nums[i] = 1 and i = i+1

i = 2, j = 2, nums[i] = 2, nums[j] = 2 => do nothing

i = 2, j = 3, nums[i] = 2, nums[j] = 2 => do nothing

i = 2, j = 4, nums[i] = 2, nums[j] = 3 => nums[j]!=2 => set nums[i] = 3 and i = i+1

i = 3, j = 5, nums[i] = 2, nums[j] = 0 => nums[j]!=2 => set nums[i] = 0 and i = i+1

i = 4, j = 6, nums[i] =3, nums[j] = 4 => nums[j]!=2 => set nums[i] = 4 and i = i+1

i = 5, j = 7, nums[i] = 3, nums[j] = 2 => do nothing

END LOOP 
We have that nums = [0, 1, 3, 0, 4, 0, 4, 2]
return k
```

# Complexity

We traverse the array 1 time, so time complexity is $O(n)$. The space complexity is $O(1)$.

````python
def removeElement(nums, val):
  i = 0
  for j in range(len(nums)):
    if nums[j] != val:
      nums[i] = nums[j]
      i+=1
  return i
````
