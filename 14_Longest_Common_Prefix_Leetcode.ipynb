# 14. Longest Common Prefix

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string `""`.

 
## Example 1:

```
Input: strs = ["flower","flow","flight"]
Output: "fl"
```

## Example 2:

```
Input: strs = ["dog","racecar","car"]
Output: ""
Explanation: There is no common prefix among the input strings.
``` 

## Constraints:
```
1 <= strs.length <= 200
0 <= strs[i].length <= 200
strs[i] consists of only lowercase English letters
```

# Solution

We solve this by creating a while loop and within the while loop we create a for loop. The while loop is to track the letter index we're tracking and the for loop is for us to loop through each string in the strs list and check whether each letter is equal to the letter of the first string. Example: Our list of strings is `["flower","flow","flight"]`. We start of with `j=0`. This means that in the for loop, we'll be checking whether the strings after the first string have letters at index `0` equal to `f`, which is the letter at index `0` for the first string. We check, and indeed they are equal, so we add that to the list of `common_strings` and move on to check whether all of the strings have the same letter at index `1`, and so on until we have a mismatch and break out of both loops.


# Complexity
The time complexity is $O(nm)$ since in each iteration of the while loop, we iterate throught the list of strings via for loop. Whilst the for loop is $O(n)$, we do the for loop for each letter of the first string, until there is a mismatch. This is $O(m)$. Hence, the overal time complexity is $O(m)$. The space complexity is $O(m)$ since we create a list which stores the common strings.

````python
def longestCommonPrefix(strs):
        common_string = []

        j = 0
        e = True
        while e and j < len(strs[0]):
            first = strs[0][j]
            for i in range(1,len(strs)):
                  if j == len(strs[i]) or strs[i][j] != first:
                        
                        e = False
                        break
            j += 1
            if e == True:
                  common_string.append(first)
        return ''.join(common_string) if common_string else ""
````
