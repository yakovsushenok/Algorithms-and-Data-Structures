# 118. Pascal's Triangle

Given an integer numRows, return the first numRows of Pascal's triangle.

## Example 1:

```
Input: numRows = 5
Output: [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]
Outout:
[    [1],
    [1,1],
   [1,2,1],
  [1,3,3,1],
 [1,4,6,4,1]]
```

## Example 2:

```
Input: numRows = 1
Output: [[1]]
```

## Constraints:

```
1 <= numRows <= 30
```

# Solution
This solution will be utilizing the concept of dynamic programming. Why? Because we'll be using the past row to update the current row. I'll explain the steps of the algorithm. We first start by checking if `numRows` is equal to `1`. If it is, we just return `1`. If not, we continue. Now, I initialize my `pascal_triangle` to `[[1],[1,1]]`, so that I can start summing up the elements for the next row. Now, I will loop from `2` to `numRows` to append those rows to `pascal_triangle`. Inside the loop, I will create a `prev_row`, which stands for the previous iteration's row array, and I will create a `current_row`, which is the row array I want to append in this iteration. I will create the current row array by initializing it to be an array of `1`'s with length `row+1` (remember we started `row` from `0` so the array on `row=0` will have length `1` => the array at `row=i` will have length `i+1`). I do this because The first and last elements will be `1`'s and I will just change the ones in the middle. I change them by creating a loop and looping though the current row array and the previous row array. For the current one, I loop from indices `j=1` to `j=length(current_row)-2`. So if the length of the current row is `4`, then I loop from index `j=1` to `j=2`. And I set them to be equal to the previous row array's sum of consecutive elements. So for example if the the current row is `4`, then the prevous row is `3`. The index at `j=1` for the current row will equal `prev_row[0] + prev_row[1] = 1+2=3`. We do this 2 times.

# Complexity
The time complexity is $O(n^2)$ where `n=numRows`. This is because for each row, we iterate through an array dependent of length equal to the row's number. 

The space complexity in my algorithm is $O(n)$, because in each iteration I create an array of length equal to the row's number. I can make the space complexity equal to $O(1)$ by just setting the `current_row[j] = pascal_triangle[row-1][j-1] + pascal_triangle[row-1][j]`.

````python
def generate(numRows):
    """
    :type numRows: int
    :rtype: List[List[int]]
    """
    if numRows == 1:
      return [[1]]
    else:
      pascal_triangle = [[1], [1,1]]
      for row in range(2,numRows):
        prev_row = pascal_triangle[row-1]
        current_row = [1]*(row+1)
        for j in range(1,len(current_row)-1):
          current_row[j] = prev_row[j-1] + prev_row[j]
        pascal_triangle.append(current_row)
      return pascal_triangle
````
