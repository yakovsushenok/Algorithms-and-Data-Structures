# Insertion Sort Algorithm

```
i ← 1
while i < length(A)
    j ← i
    while j > 0 and A[j-1] > A[j]
        swap A[j] and A[j-1]
        j ← j - 1
    end while
    i ← i + 1
end while
```

````python
def insertionSort(nums):
  i = 1
  while i < len(nums):
    j = i
    while j > 0 and nums[j-1] > nums[j]:
      nums[j], nums[j-1] = nums[j-1], nums[j]
      j -= 1
    i += 1
  return(nums)

print(f"Array {[2,6,1,4]} Sorted: {insertionSort([2,6,1,4])}")
print(f"Array {[3,9,0,-2, -2, -3]} Sorted {insertionSort([3,9,0,-2, -2, -3])}")
````
