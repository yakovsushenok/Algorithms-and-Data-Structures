Given an integer `rowIndex`, return the `rowIndexth` (0-indexed) row of the Pascal's triangle.

In Pascal's triangle, each number is the sum of the two numbers directly above it as shown:

## Example 1:

```
Input: rowIndex = 3
Output: [1,3,3,1]
```
## Example 2:

```
Input: rowIndex = 0
Output: [1]
```
## Example 3:

```
Input: rowIndex = 1
Output: [1,1]
``` 

## Constraints:

```
0 <= rowIndex <= 33
``` 

Follow up: Could you optimize your algorithm to use only `O(rowIndex)` extra space?

# Solution: TODO (Explanation)



````python
def getRow(rowIndex):
    """
    :type rowIndex: int
    :rtype: List[int]
    """
    # 0 :     [1]
    # 1 :    [1,1]
    # 2 :   [1,2,1]
    # 3 :  [1,3,3,1]
    # 4 : [1,4,6,4,1]
    # 5 :[1,5,10,10,5,1]
    if rowIndex == 0:
        return [1]
    pasc_triangle = [[1],[1,1]]
    for i in range(2,rowIndex+1):
        prev_row = pasc_triangle[i-1]
        pasc_triangle.append([0]*(i+1))
        pasc_triangle[i][0], pasc_triangle[i][-1]  = 1, 1
        for j in range(1, i):
            pasc_triangle[i][j] = prev_row[j-1] + prev_row[j]
    
    return pasc_triangle[rowIndex]
````
