# 13. Roman to Integer

Roman numerals are represented by seven different symbols: ```I```, ```V```, ```X```, ```L```, ```C```, ```D``` and ```M```.

```
Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
```

For example, ```2``` is written as ```II``` in Roman numeral, just two ones added together. ```12``` is written as `XII`, which is simply `X + II`. The number `27` is written as `XXVII`, which is `XX + V + II`.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not `IIII`. Instead, the number four is written as `IV`. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as `IX`. There are six instances where subtraction is used:

`I` can be placed before `V` (5) and `X` (10) to make 4 and 9. 
`X` can be placed before `L` (50) and C (100) to make 40 and 90. 
`C` can be placed before `D` (500) and M (1000) to make 400 and 900.


Given a roman numeral, convert it to an integer.


## Example 1:
```
Input: s = "III"
Output: 3
Explanation: III = 3.
```

## Example 2:
```
Input: s = "LVIII"
Output: 58
Explanation: L = 50, V= 5, III = 3.
```

## Example 3:

```
Input: s = "MCMXCIV"
Output: 1994
Explanation: M = 1000, CM = 900, XC = 90 and IV = 4.
```

## Constraints:

```
1 <= s.length <= 15
s contains only the characters ('I', 'V', 'X', 'L', 'C', 'D', 'M').
It is guaranteed that s is a valid roman numeral in the range [1, 3999].
```

# Solution

I'm going to solve this problem by traversing over the roman numeral string one time. The way I'm going to do it is the following:
```
Initialize subtract = True    # this is a boolen variable which I'm going to use to check if we encountered a roman 
numeral of the sort IV, where I need to subtract I from V to get 4. Basically at every numeral I check if the next numeral is 
bigger than the previous. If it is, then we do the subtraction (e.g. as above) and we skip the next numeral.

Initialize sumInt = 0 # this will be my sum of the numerals
We now iterate through all of the roman numerals. At each iteration we do the following

1.   Check if the index of the element is the last. I do this because I'm considering the case where at the previous iteration we were at the second last index and there was a case of subtraction (i.e. the second last element was I and the last element was V), then we would finish the loop. If there was no subtraction, then just add the last element.
2. In the case that we're not at the last element, we have two possibilities:
2.1 If subtract is true, skip the current index
2.2 If subtract is false, check if the next element of the list is bigger. If:
2.2.1 the next element is larger -- get the subtraction amount (i.e. V - I = 4) and add it to the sum. Set subtract equal to True.
2.2.2 the next element is not larger -- add the current element to the sum


We do this until we traverse the whole list.

```

````python
def romanToInt(s):
  s = list(s)
  romanDict = {"I" : 1, "V" : 5, "X" : 10, "L" : 50, "C" : 100, "D" : 500, "M" : 1000}
  sumInt = 0
  subtract = False
  for ind, i in enumerate(s):
      if ind < len(s) - 1:
          if subtract == True:
              subtract = False
              continue
          else:
              if romanDict[s[ind]] < romanDict[s[ind+1]]:
                  summand = romanDict[s[ind+1]] - romanDict[s[ind]]
                  sumInt += summand
                  subtract = True
              else:
                  sumInt += romanDict[i]
      else:
          if subtract == True:
              break
          else:
              sumInt += romanDict[i]
  return sumInt
````
