# 88. Merge Sorted Array

You are given two integer arrays `nums1` and `nums2`, sorted in non-decreasing order, and two integers `m` and `n`, representing the number of elements in `nums1` and `nums2` respectively.

Merge `nums1` and `nums2` into a single array sorted in non-decreasing order.

The final sorted array should not be returned by the function, but instead be stored inside the array `nums1`. To accommodate this, `nums1` has a length of `m + n`, where the first `m` elements denote the elements that should be merged, and the last `n` elements are set to `0` and should be ignored. `nums2` has a length of `n`.

## Example 1:
```
Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
Output: [1,2,2,3,5,6]
Explanation: The arrays we are merging are [1,2,3] and [2,5,6].
The result of the merge is [1,2,2,3,5,6] with the underlined elements coming from nums1.
```

## Example 2:

```
Input: nums1 = [1], m = 1, nums2 = [], n = 0
Output: [1]
Explanation: The arrays we are merging are [1] and [].
The result of the merge is [1].
```

## Example 3:

```
Input: nums1 = [0], m = 0, nums2 = [1], n = 1
Output: [1]
Explanation: The arrays we are merging are [] and [1].
The result of the merge is [1].
Note that because m = 0, there are no elements in nums1. The 0 is only there to ensure the merge result can fit in nums1.
```

## Constraints:

```
nums1.length == m + n
nums2.length == n
0 <= m, n <= 200
1 <= m + n <= 200
-109 <= nums1[i], nums2[j] <= 109
```

##### Follow up: Can you come up with an algorithm that runs in O(m + n) time?

# Easy-level Solution
The easy level solution is to create a copy of the `nums1` first m elements, `nums1c`, and then for each index of `nums1` to pick the smaller element of the 2 arrays - `nums1c` and `nums2`. This is a 3 pointer technique. We will have `i` to iterate over `nums1c`, then `j` to iterate over `nums2` and `k` for `nums1`. At each `k`, we compare whether `nums1c[i] < nums2[j]` or not. If it's true, then we set `nums1[k]` to be equal to `nums1c[i]` and we increment both `i` and `k` by `1`. If the condition is not true, we do the same but for `j`. Note that we do this only until `k` is equal to the length of `nums1` or until we get `i` to equal `m` or `j` to equal `i`. If `i` is equal to `m`, then we know that we have no more elements to take from `nums1c`, so we iterate over the left over elements from `nums2`. Same for `j=n`. 

# Complexity 
This is an easy level solution because we create an array of length `m`, thus using space. The medium level solution doesn't use any additional space. The time complexity of this solution is $O(m+n)$. The space complexity is $O(m)$.

````python
def merge(nums1, m, nums2, n):
  if  m >= 0 and n == 0:
    nums1 = nums1
  elif m == 0 and n > 0:
    for i in range(n):
      nums1[i] = nums2[i]
  else:
    i = 0
    j = 0
    k = 0
    nums1c = nums1.copy()[:m]
    while k < n + m:
      if j == n:
        for s in range(i, m):
          nums1[n+s] = nums1c[s]
        break
      if i == m:
        for s in range(j, n):
          nums1[m+s] = nums2[s]
        break
      if nums1c[i] < nums2[j]:
          nums1[k] = nums1c[i]
          i+=1
          k+=1
      else:
          nums1[k] = nums2[j]
          j+=1
          k+=1
````

# Medium-level solution
This is a medium-level solution because it doesn't use any additional space (i.e. it's space complexity is $O(1)$ - constant. I'll explain the algorithm - it's a 3 pointer technique like the previous one, however this time we're going to start from the end with `k` going from `m + n - 1` to `0` and the indices `i` and `j` from `m` and `n` to `0` respectively. At each iteration, we compare which is the bigger number from `nums1[i]` and `nums2[j]` and set it to `nums1[k]`. If `nums1[i] > nums2[j]` then we subtract `1` from both `i` and `k`. We do the same for `j` if `nums1[i] =< nums2[j]`. Notice that we put an additional condition in the `nums1[i] > nums2[j]` condition, that is `if i >= 0`. This way if `i` is smaller than `0` and `j` is still larger than `-1`, this means that we haven't appended all of the elements of `j` to `nums1`. We also put an additional condition to the while loop - `if j > 0`. We break the while loop if `j` is smaller than `0`. Think about it, if we iterated through all of the elements from `nums2`, this means that all of the elements from `nums1` are smaller, and those elements are already in order in `nums1`.

# Complexity
Time complexity: $O(m+n)$

Space complexity: $O(1)$ - we don't use any space

````python
def merge(nums1, m, nums2, n):
    if  m >= 0 and n == 0:
      nums1 = nums1
    elif m == 0 and n > 0:
      for i in range(n):
        nums1[i] = nums2[i]
    else:
      i, j, k = m - 1, n - 1, m + n - 1
      while k > -1:
        if j < 0:
          break
        if i >= 0 and nums1[i] > nums2[j]:
          nums1[k] = nums1[i]
          k -= 1
          i -= 1
        else:
          nums1[k] = nums2[j]
          k -= 1
          j -= 1
````
