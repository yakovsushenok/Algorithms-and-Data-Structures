# 26. Remove Duplicates from Sorted Array

Given an integer array `nums` sorted in **non-decreasing order**, remove the duplicates **in-place** such that each unique element appears only **once**. The **relative order** of the elements should be kept the **same**.

Since it is impossible to change the length of the array in some languages, you must instead have the result be placed in the **first part** of the array `nums`. More formally, if there are `k` elements after removing the duplicates, then the first `k` elements of `nums` should hold the final result. It does not matter what you leave beyond the first `k` elements.

Return `k` after placing the final result in the first `k` slots of `nums`.

Do **not** allocate extra space for another array. You must do this by **modifying the input array in-place** with O(1) extra memory.

# Solution

Since we need to solve this without creating a second array, i.e. in-place, I will use a two pointer strategy. The first pointer `i` will be referencing only the first `k` (number of non-duplicate entries) in the `nums` array, and the second pointer `j` will be going through the whole array. When `j` encounters an element which is different than the current `i`th element, we make the `i+1`th element equal to the `j`th element and then increment `i` by 1. Example:

```
let nums = [0,0,1,1,1,2,2,3,3,4]
Initialize i = 0 and j = 1
loop for j in [1, length(nums)]:

i = 0, j = 1, nums[i] = 0, nums[j] = 0 => nums[i] = nums[j] => don't do anything => nums = [0,0,1,1,1,2,2,3,3,4]

i = 0, j = 2, nums[i] = 0, nums[j] = 1 => nums[i] != nums[j] => nums[i+1] = 1 and i = i + 1 and 
nums = [0,1,1,1,1,2,2,3,3,4]

i = 1, j = 3, nums[i] = 1, nums[j] = 1 => don't do anything

i = 1, j = 4, nums[i] = 1, nums[j] = 1 => don't do anything

i = 1, j = 5, nums[i] = 1, nums[j] = 2 => nums[i+1] = 2 => nums = [0,1,2,1,1,2,2,3,3,4] and i = i + 1

i = 2, j = 6, nums[i] = 2, nums[j] = 2 => don't do anything

i=2, j = 7, nums[i] = 2, nums[j] = 3 => nums[i+1] = 3 => nums = [0,1,2,3,1,2,2,3,3,4] and i = i + 1

i = 3, j = 8, nums[i] = 3, nums[j] = 3 => don't do anything

i = 3, j = 9, nums[i] = 3, nums[j] = 4 => nums[i+1] = 4 => nums = [0,1,2,3,4,2,2,3,3,4]

END LOOP

return i + 1 # we return i+1 since i started counting the number of non-duplicate numbers from 0
```



# Complexity

We traverse the array 1 time, hence the complexity is $O(n)$. The space complexity is $O(1)$.

````python
def removeDuplicates(nums):
  i = 0
  for j in range(1, len(nums)):
    if nums[i] != nums[j]:
      nums[i+1] = nums[j]
      i+=1
  return i + 1   
````
