# 66. Plus One

You are given a large integer represented as an integer array `digits`, where each `digits[i]` is the `i`th digit of the integer. The digits are ordered from most significant to least significant in left-to-right order. The large integer does not contain any leading `0`'s.

Increment the large integer by one and return the resulting array of digits.

 

## Example 1:

```
Input: digits = [1,2,3]
Output: [1,2,4]
Explanation: The array represents the integer 123.
Incrementing by one gives 123 + 1 = 124.
Thus, the result should be [1,2,4].
```

## Example 2:

```
Input: digits = [4,3,2,1]
Output: [4,3,2,2]
Explanation: The array represents the integer 4321.
Incrementing by one gives 4321 + 1 = 4322.
Thus, the result should be [4,3,2,2].
```

## Example 3:

```
Input: digits = [9]
Output: [1,0]
Explanation: The array represents the integer 9.
Incrementing by one gives 9 + 1 = 10.
Thus, the result should be [1,0].
```
```
Constraints:

1 <= digits.length <= 100
0 <= digits[i] <= 9
digits does not contain any leading 0's.
```

# Solution

We can do it 2 ways: 

1. Converting the list to strings -> converting back to integer after joining -> incrementing by 1 to the integer -> converting back to string -> creating a list from the numbers.
2. By not converting to strings.

### Solution 1

This solution is pretty easy. First, I create a list of strings which are the string versions of the integers in `digits`. After that, I `join` them and get 1 string. I then turn it into an integer and after that I add 1. After that, I do the inverse of what I've done. First I turn the incremented integer into a string and then after that I create a list from the integers of the string.

### Complexity 

The time complexity is $O(n)$ as we do 3 loops which are $O(n)$. So $O(3n) = O(n)$. Space complexity is $O(n)$ since we create a new $n+1$ array.

### Solution 2

In this solution we will do 2 loops. In the first loop, we will get the integer from the list. We will do that by summing the numbers multiplied by their respective powers of 10. So for example if we have the following list: `[4,3,2,1]`, we get the number with the following sum: `4*1000 + 3*100 + 2*10 + 1*1 = 4321`. Now we just increment the sum and extract the numbers. We do that in a loop. In an iteration, we get the last digit by getting modulo of 10, then for the next iteration we prepare the number by getting rid of that last number by just dividing by 10 and converting the `flloat` into an `int`. Also note that we are using `//` instead of `/` as the latter messes up the conversion. 

### Complexity 

The time complexity is $O(n)$ as we do 2 loops which are $O(n)$. So $O(2n) = O(n)$. Space complexity is $O(n)$ since we create a new $n+1$ array.

# Solution 1

````python
def plusOne(digits):
    """
    :type digits: List[int]
    :rtype: List[int]
    """
    c = [str(i) for i in digits]
    d = ''.join(c)
    sum = int(d) + 1 
    s = str(sum)
    return [int(i) for i in s]
````

# Solution 2

````python
def plusOne(digits):
  pow = 0
  sum = 0
  for i in range(len(digits)-1, -1, -1):
    multiplier = 10**pow
    pow+=1
    sum += digits[i]*multiplier
  sum += 1 
  newDigits = []
  while True:
    if sum < 10:
      newDigits.append(sum)
      break
    newDigits.append(sum%10)
    sum = int(sum//10)
  newDigits = newDigits[::-1]
  return newDigits
````
