Given a stream of data representing API calls, design a data structure that can track the latency and errors within the applications.

Implement the `MonitoringSystem` class:

  * `MonitoringSystem()` Initializes the object.
  * `void logLatency(String applicationName, String api, int latencyInMills)` Records an API call `api` for an application `applicationName` which took `latencyInMills` milliseconds to execute.
  * `void logError(String applicationName, String api, int errorCode)` Records an error log `errorCode` that happened for the given application and API combination. Note that two different application and API combinations may have the same error code.
  * `int getPercentileLatency(int percentile, String applicationName, String api)` Returns the percentile latency of a given application and API combination. In case there are no API calls with the given combination, it returns the percentile latency of all API calls for the given application.
Percentile latency is the latency value that is greater than or equal to percentile percent of all latency values. For example, the 95th percentile latency is the latency value that is greater than or equal to `95%` of all latencies.
More formally, percentile latency is equal to `latency[index]` where latency is a sorted list of all latencies for the given combination, and `index = max(0, ⌊(percentile * latency.length) / 100⌋ - 1)`.
  * `int[] getTopErrors(String applicationName, String api)` Returns the 3 most frequent error codes for the given application and API combination. The returned error codes should be sorted in decreasing order by number of occurrences. If two error codes occur the same number of times, the numerically smaller error code should come first.
In case there are less than `3` error codes with the given combination, it returns the error codes of the 3 most frequent error logs of all the error logs under the given application. Note that in this case, the same error code may occur more than once in the answer.
For example, say there are four error logs: `("app", "a", 404)`, `("app", "a", 404)`, `("app", "b", 404)`, and `("app", "b", 100)`. Then, `[404, 100, 404]` is returned for the application `"app"`.

Explanation
MonitoringSystem ms = new MonitoringSystem();
ms.logLatency("profile", "getuser", 100); // the "getuser" API of "profile" application took 100ms to execute.
ms.logLatency("profile", "getuser", 100); // same as above.
ms.logError("profile", "updateuser", 400); // the "updateuser" API of "profile" generated error 400.
ms.logLatency("profile", "updateuser", 200); // the "updateuser" API of "profile" took 200ms to execute.
ms.logError("profile", "updateuser", 504); // the "updateuser" API of "profile" generated error 504.
ms.logError("profile", "updateuser", 504); // same as above.

## Constraints:

```
1 <= applicationName.length <= 10
1 <= api.length <= 10
1 <= latencyInMills <= 106
1 <= errorCode <= 106
95 <= percentile <= 99
applicationName and api consist of lowercase English letters only.
At most 1000 calls in total will be made to logLatency, logError, getPercentileLatency, and getTopErrors.
At least one call will be made to getPercentileLatency or getTopErrors.
getPercentileLatency will be called when there is at least one API call with applicationName.
getErrors will be called when there are at least three distinct errors under applicationName.
```

````python
class MonitoringSystem(object):

    def __init__(self):

        self.apps = {}

    def logLatency(self, applicationName, api, latencyInMills):
        """
        :type applicationName: str
        :type api: str
        :type latencyInMills: int
        :rtype: None
        """
        
        if applicationName not in self.apps:
            self.apps[applicationName] = {api: {"latency": [latencyInMills]}} 
        else:
            if api not in self.apps[applicationName]:
                self.apps[applicationName][api] = {"latency": [latencyInMills]}
            else:
                if "latency" not in self.apps[applicationName][api]:
                    self.apps[applicationName][api]['latency'] =  [latencyInMills]
                else:
                    self.apps[applicationName][api]['latency'].append(latencyInMills)
        

    def logError(self, applicationName, api, errorCode):
        """
        :type applicationName: str
        :type api: str
        :type errorCode: int
        :rtype: None
        """
        if applicationName not in self.apps:
            self.apps[applicationName] = {api :{"errorCode":{errorCode:1}}}
        else:
            if api not in self.apps[applicationName]:
                self.apps[applicationName][api] = {"errorCode":{errorCode:1}}
            else:
                if "errorCode" in self.apps[applicationName][api]:
                    if errorCode in self.apps[applicationName][api]['errorCode']:
                        self.apps[applicationName][api]['errorCode'][errorCode] += 1
                    else:
                        self.apps[applicationName][api]['errorCode'][errorCode] = 1
                else:
                    self.apps[applicationName][api]['errorCode'] = {errorCode:1}

            

    def getPercentileLatency(self, percentile, applicationName, api):
        """
        :type percentile: int
        :type applicationName: str
        :type api: str
        :rtype: int
        """
        if api in self.apps[applicationName] and "latency"  in self.apps[applicationName][api]:
            l = len(self.apps[applicationName][api]['latency'])
            latencyList = sorted(self.apps[applicationName][api]['latency'])
            index = max(0, int(percentile * l / 100) - 1)
            return latencyList[index]
        else:
            latencyList = []
            for api in self.apps[applicationName]:
                if "latency" not in self.apps[applicationName][api]:
                    continue
                latencyList.extend(self.apps[applicationName][api]['latency'])
            l = len(latencyList)
            latencyList = sorted(latencyList)
            index = max(0, int(percentile * l / 100) - 1)
            return latencyList[index]


    def getTopErrors(self, applicationName, api):
        """
        :type applicationName: str
        :type api: str
        :rtype: List[int]
        """

        if api in self.apps[applicationName]:
            if "errorCode" in self.apps[applicationName][api]:
                if len(self.apps[applicationName][api]["errorCode"]) >= 3:

                        errorDict = self.apps[applicationName][api]["errorCode"]
                        sortedVals = sorted(errorDict.items(),key = lambda x: (-x[1], x[0]))

                        final3codes = [i[0] for i in sortedVals[:3]]
                        return final3codes
        
        lst = []
        for api in self.apps[applicationName]:
            if "errorCode" not in self.apps[applicationName][api]:
                continue
            for error in self.apps[applicationName][api]['errorCode']:
                lst.append([api, error, self.apps[applicationName][api]['errorCode'][error]])
        lstSorted = sorted(lst, key = lambda x: (-x[2], x[1]))
        final3codes = [i[1] for i in lstSorted[:3]]
        return final3codes
````
